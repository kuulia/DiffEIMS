```python
from rdkit import Chem
from rdkit.Chem import Draw

# Explicit H SMILES for radicals
isoprene = "CC(=C)C=C"                # starting molecule
ro_radical = "[CH2]C(O)C=C"           # RO* radical
ro2_radical = "[CH2]C(OO)C=C"         # RO2 radical (simplified hydroperoxide)
peroxide = "CC(O)C(OO)C"               # closed-shell peroxide

mol_steps = [isoprene, ro_radical, ro2_radical, peroxide]
step_names = ["Isoprene", "RO* radical", "RO2 radical", "ROOR peroxide"]

mols = [Chem.MolFromSmiles(sm) for sm in mol_steps]
img = Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(300,300), legends=step_names)
img
```
